```python
import os

def save_uploaded_audio(uploader):
    for filename, file_info in uploader.value.items():
        with open(filename, 'wb') as f:
            f.write(file_info['content'])
        print(f"✅ Saved: {filename}")
        return filename
    return None

# Call the function after uploading
audio_filename = save_uploaded_audio(uploader)
```

```python
import requests

# Configuration
API_URL = "https://api.sarvam.ai/speech-to-text-translate"
API_KEY = "c29010c0-9f68-4567-adf3-30d5cbb5d9e0"  # Replace with your actual key
AUDIO_FILE_PATH = "Cessna_Business_Park.m4a"  # Replace with your actual audio file

def transcribe_audio(audio_path):
    headers = {
        "api-subscription-key": API_KEY
    }

    # Open audio file
    with open(audio_path, "rb") as audio_file:
        files = {
            "file": (audio_path, audio_file, "audio/wav")
        }

        response = requests.post(API_URL, headers=headers, files=files)

    if response.status_code == 200:
        print("✅ Transcription Success!")
        print("Transcribed Text:", response.json())
    else:
        print("❌ API Error:", response.status_code)
        print("Details:", response.text)

if __name__ == "__main__":
    transcribe_audio(AUDIO_FILE_PATH)
```
